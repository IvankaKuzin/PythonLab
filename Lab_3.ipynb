```python
n = int(input('Введіть N= '))
for i in range(n,-1,-1):
  for j in range(0,i):
    print(n, end=' ')
  print("\r") 
for k in range(0,n):
  for l in range(0,k+1):
    print(n, end=' ')
  print("\r") 
```

```python
s1=list(input("Слово 1: "))
s2=list(input("Слово 2: "))
s3=list(input("Слово 3: "))
nepowt=[]
powt=[]
i=0
j=0
k=0
for i in range(0,len(s1)):
  for j in range(0,len(s2)):   
    for k in range(0,len(s3)): 
      if (s1[i] in s2)or (s1[i]in s3):
        powt.append(s1[i])
      elif (s2[j] in s1) or (s2[j]in s3):
          powt.append(s2[j])
      elif (s3[k] in s1)or (s3[k]in s2):
            powt.append(s3[k])
      else:
            nepowt.append(s1[i])
            nepowt.append(s2[j])
            nepowt.append(s3[k])
else:
  print("Символ nepovtor", set(nepowt))
  print("Символ povtor", set(powt))
```

```python
from math import sqrt
n=int(input('Введіть граничне значення: '))
k = 0
sum = 0
dob = 1
while k < n:
  k = k + 1
  sum = sum + (1/k+sqrt(k))
  dob = dob * (1/k+sqrt(k))
  
else:
  print('Сума', n, 'значень =', sum)
  print('Добуток', n, 'значень =', dob)
```
