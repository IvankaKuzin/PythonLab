```python
from random import randint
import math
n=int(input("Введіть кількість елементів: "))
mas=[randint(-10,10) for x in range(n)]
print("Масив рандомних чисел: ",mas)

for i in range(n):
  if mas[i]<0:
    mas[i]=abs(mas[i])
print("Числа масиву по модулю: ",mas)

mas.remove(max(mas))
mas.remove(min(mas))
print("Масив без мах і min значень по модулю: ",mas)

dob=1
for i in range(n-2):  
  dob*=abs(mas[i])
print("Добуток елементів= ",dob)
```
